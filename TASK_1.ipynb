## Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
```

## Loading the dataset

```python
file_path = 'world_population.csv'  
data = pd.read_csv(file_path)
```

## Displaying first few rows

```python
data.head()
```

## Checking the structure of Data

```python
data.info()
```

## Summary statistics of numeric columns

```python
data.describe()
```

## Sort by population and get top 10 countries

```python
top_countries = data.sort_values(by="World Population Percentage", ascending=False).head(10)
```

## Plot the top 10 populous countries

```python
plt.figure(figsize=(12, 6))
sn.barplot(x="Country", y="2022 Population", data=top_countries)
plt.title("Top 10 Most Populous Countries in 2022")
plt.xlabel("Country")
plt.ylabel("2022 Population")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Group data by continent and sum population for 2022

```python
continent_population = data.groupby("Continent")["2022 Population"].sum().reset_index()
```

## Plot population distribution by continent in 2022

```python
plt.figure(figsize=(10, 6))
sn.barplot(x="Continent", y="2022 Population", data=continent_population)
plt.title("Population Distribution by Continent in 2022")
plt.xlabel("Continent")
plt.ylabel("Total Population in 2022")
plt.tight_layout()
plt.show()
```

## Calculate population growth from 2020 to 2022

```python
data["Growth (2020 to 2022)"] = data["2022 Population"] - data["2020 Population"]
```

## Sort by the population growth from 2020 to 2022

```python
top_growth_countries = data.sort_values(by="Growth (2020 to 2022)", ascending=False).head(10)

plt.figure(figsize=(12, 6))
sn.barplot(x="Country", y="Growth (2020 to 2022)", data=top_growth_countries)
plt.title("Top 10 Countries by Population Growth from 2020 to 2022")
plt.xlabel("Country")
plt.ylabel("Population Growth (2020 to 2022)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Sort by population density in 2022 and get the top 10 countries

```python
top_density = data.sort_values(by="Density (per km²)", ascending=False).head(10)

plt.figure(figsize=(12, 6))
sn.barplot(x="Country", y="Density (per km²)", data=top_density)
plt.title("Top 10 Countries by Population Density in 2022")
plt.xlabel("Country")
plt.ylabel("Density (per km²)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Group by continent and calculate the total world population percentage for each continent in 2022

```python
continent_population_percentage = data.groupby("Continent")["World Population Percentage"].sum().reset_index()

plt.figure(figsize=(10, 6))
sn.barplot(x="Continent", y="World Population Percentage", data=continent_population_percentage)
plt.title("World Population Percentage by Continent in 2022")
plt.xlabel("Continent")
plt.ylabel("World Population Percentage")
plt.tight_layout()
plt.show()
```

## Plot distribution of population growth rate in 2022

```python
plt.figure(figsize=(10, 6))
sn.histplot(data["Growth Rate"], bins=30, kde=True, color="blue")
plt.title("Population Growth Rate Distribution in 2022")
plt.xlabel("Growth Rate (%)")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

## Area vs. Population (2022)

```python
plt.figure(figsize=(10, 6))
sn.scatterplot(x="Area (km²)", y="2022 Population", data=data, color="green")
plt.title("Area vs. Population in 2022")
plt.xlabel("Area (km²)")
plt.ylabel("2022 Population")
plt.tight_layout()
plt.show()
```
